```python
import os
from PIL import Image

def is_image_file_valid(image_path):
    try:
        with Image.open(image_path) as img:
            img.verify()
        return True
    except (IOError, SyntaxError):
        return False

def delete_invalid_image_dirs(root_dir):
    for root, dirs, _ in os.walk(root_dir):
        print(root)
        print(dirs)
        for dir in dirs:
            print(dir)
            dir_path_parent = os.path.join(root, dir)
            for file in os.listdir(dir_path_parent):
                dir_path_child = os.path.join(dir_path_parent, file)
                for file in os.listdir(dir_path_child):
                    if file.lower().endswith('.jpg'):
                        file_path = os.path.join(dir_path_child, file)
                        print(is_image_file_valid(file_path))
                        if not is_image_file_valid(file_path):
                            print(f"Deleting invalid image directory: {dir_path_child}")
                            # os.rmdir(dir_path)
                            break
        break
        

# 사용법
delete_invalid_image_dirs('/home/cschoi/VR/dataset/Classification/task1/category1/test')
```
